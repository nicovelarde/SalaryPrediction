```python
import matplotlib.pyplot as plt
import pandas as pd
import math
import numpy as np

# Load and visualize data
file = pd.read_csv('salary_data.csv')
x = file['YearsExperience'].tolist()
y = file['Salary'].tolist()
plt.scatter(x, y)
plt.title('Years of Experience vs Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.savefig('plot_matplotlib.png')
plt.show()

# Calculate mean of X and Y
x_mean = np.mean(x)
y_mean = np.mean(y)

# Calculate errors (deviations from the mean)
error_x = [value - x_mean for value in x]
error_y = [value - y_mean for value in y]

# Calculate the numerator for B1
errors = [error_x[i] * error_y[i] for i in range(len(error_x))]
errors_sum = sum(errors)

# Calculate the denominator for B1
sum_sq_error_x = sum([i**2 for i in error_x])

# Calculate B1 and B0
b1 = errors_sum / sum_sq_error_x
b0 = y_mean - b1 * x_mean

# Predict Y values
Predicted_y = [b0 + b1 * i for i in x]

# Plot the results
plt.scatter(x, y, color='blue', label='Actual Data')
plt.plot(x, Predicted_y, linestyle='dashed', color='orange', label='Regression Line')
plt.title('Years of Experience vs Salary with Regression Line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.savefig('regression_plot_prediction_matplotlib.png')
plt.show()

# Calculate the prediction error
differences = [Predicted_y[i] - y[i] for i in range(len(y))]
square_errors = [diff**2 for diff in differences]

# Calculate Mean Squared Error (MSE)
mse = sum(square_errors) / len(y)

# Calculate Root Mean Squared Error (RMSE)
rmse = math.sqrt(mse)

# Calculate R
ss_total = sum([(value - y_mean)**2 for value in y])
ss_residual = sum(square_errors)
r2 = 1 - (ss_residual / ss_total)

print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
print(f'Coefficient of Determination (R): {r2}')

```
