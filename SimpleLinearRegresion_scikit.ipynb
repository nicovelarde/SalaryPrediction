```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score

# Load data
file = pd.read_csv('salary_data.csv')
x = file['YearsExperience']
y = file['Salary']

# Create a DataFrame for Seaborn
data = pd.DataFrame({
    'YearsExperience': x,
    'Salary': y
})

# Plot the data and regression line using Seaborn
plt.figure(figsize=(10, 6))
sns.regplot(x='YearsExperience', y='Salary', data=data, ci=None, scatter_kws={'color': 'blue'}, line_kws={'color': 'orange', 'linestyle': 'dashed'})
plt.title('Years of Experience vs Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.savefig('regression_plot_seaborn.png')
plt.show()

# Calculate the regression coefficients
x_mean = np.mean(x)
y_mean = np.mean(y)
error_x = x - x_mean
error_y = y - y_mean

b1 = sum(error_x * error_y) / sum(error_x**2)
b0 = y_mean - b1 * x_mean

# Make predictions
y_pred = b0 + b1 * x

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y, y_pred))

# Calculate R²
r2 = r2_score(y, y_pred)

print(f'Root Mean Squared Error (RMSE): {rmse}')
print(f'Coefficient of Determination (R²): {r2}')
```
